```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as seabornInstance
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
# Readig Data
dataset = pd.read_csv("C:\\Benion\\Benion Programmings\\Python\\AI & ML\\data\\weatherHistory.csv")
```

```python
# Shape of the dataset
print(dataset.shape)
```

```python
# Description of the dataset
print(dataset.describe())
```

```python
# Linear Graph of Temperature vs Apparent Temperature
dataset.plot(x="Temperature (C)", y="Apparent Temperature (C)", style="o")
plt.title("Temperature vs Apparent Temperature")
plt.xlabel("Temperature")
plt.ylabel("Apparent Temperature")
plt.show()
```

```python
# Graph of Average Apparent Temperature
plt.figure(figsize=(15, 10))
plt.tight_layout()
seabornInstance.distplot(dataset["Apparent Temperature (C)"])
plt.show()
```

```python
# Data slicing
X = dataset["Temperature (C)"].values.reshape(-1, 1)
y = dataset["Apparent Temperature (C)"].values.reshape(-1, 1)

X_train, y_train, X_test, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Model training
regressor = LinearRegression()

# Training the Algorithm
regressor.fit(X_train, y_train)
```

```python
# Retrieve the intercept
print(f"Intercept: { regressor.intercept_ }")
```

```python
# Retrieve the coefficient
print(f"Intercept: { regressor.coef_ }")
```
