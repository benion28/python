```python
print("#-------------- Readig Community Data -------------------#")
community_data = pd.read_excel("C:\\Benion\\Benion Programmings\\Python\\AI & ML\\data\\Dummy_Communities.xlsx")
print("#-------------- Full Community Data Shape -------------------#")
community_data.shape
```

```python
print("#-------------- Readig Household Data -------------------#")
household_data = pd.read_excel("C:\\Benion\\Benion Programmings\\Python\\AI & ML\\data\\Dummy_Household.xlsx")
print("#-------------- Full Household Data Shape -------------------#")
household_data.shape
```

```python
print("#-------------- Readig Lgas Data -------------------#")
lga_data = pd.read_excel("C:\\Benion\\Benion Programmings\\Python\\AI & ML\\data\\Dummy_Lgas.xlsx")
print("#-------------- Full Lgas Data Shape -------------------#")
lga_data.shape
```

```python
print("#-------------- Extracting Community Codes -------------------#")
# Extract numeric part before "." for each code column
community_data['state_code'] = community_data['state_code'].str.extract(r'^(\d+)')
community_data['lga_code'] = community_data['lga_code'].str.extract(r'^(\d+)')
community_data['ward_code'] = community_data['ward_code'].str.extract(r'^(\d+)')

# Convert to integers (optional)
# community_data['state_code'] = community_data['state_code'].astype(int)
# community_data['lga_code'] = community_data['lga_code'].astype(int)
# community_data['ward_code'] = community_data['ward_code'].astype(int)

print("#-------------- Updated Community Dataframe -------------------#")
community_data.head(10)
```

```python
print("#-------------- Extracting Household Codes -------------------#")
# Extract numeric part before "." for each code column
household_data['code'] = household_data['code'].str.extract(r'^(\d+)')
household_data['state_code'] = household_data['state_code'].str.extract(r'^(\d+)')

# First handle NaN/blank values (make them False)
household_data['is_paid'] = household_data['is_paid'].fillna(False)
household_data['is_paid'] = household_data['is_paid'].replace('', False)

# Then convert numbers to boolean (1→True, 0→False)
household_data['is_paid'] = household_data['is_paid'].astype(bool)

print("#-------------- Updated Community Dataframe -------------------#")
household_data.head(10)
```

```python
print("#-------------- Extracting Lga Codes -------------------#")
# Extract numeric part before "." for each code column
lga_data['code'] = lga_data['code'].str.extract(r'^(\d+)')
lga_data['state_code'] = lga_data['state_code'].str.extract(r'^(\d+)')

print("#-------------- Updated Lga Dataframe -------------------#")
lga_data.head(10)
```

```python
print("#-------------- Saving Community to Excel -------------------#")

# Save to Excel file (adjust path/filename as needed)
output_community_path = "Transformed_Communities.xlsx"
community_data.to_excel(output_community_path, index=False)  # index=False prevents saving row numbers

print(f"Data successfully saved to: {output_community_path}")
```

```python
print("#-------------- Saving Household to Excel -------------------#")

# Save to Excel file (adjust path/filename as needed)
output_household_path = "Transformed_Households.xlsx"
household_data.to_excel(output_household_path, index=False)  # index=False prevents saving row numbers

print(f"Data successfully saved to: {output_household_path}")
```

```python
print("#-------------- Saving Lga to Excel -------------------#")

# Save to Excel file (adjust path/filename as needed)
output_lga_path = "Transformed_Lgas.xlsx"
lga_data.to_excel(output_lga_path, index=False)  # index=False prevents saving row numbers

print(f"Data successfully saved to: {output_lga_path}")
```
